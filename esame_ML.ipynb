```python
import numpy as np
import pandas as pd
import sklearn

from sklearn import svm
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split, StratifiedKFold
from sklearn.preprocessing import RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
from xgboost import XGBClassifier

# Configuration
dataset = pd.read_csv('train.csv')
Random_State = 23

# Utility functions
def drawCorrelationMatrix(df, title="Feature Correlation Matrix"):
    plt.figure(figsize=(16, 14))
    CM = df.corr(numeric_only=True).abs()

    sns.heatmap(CM, annot=True, fmt=".2f", cmap='Reds',
                annot_kws={"size": 6}, linewidths=0.5, linecolor='gray')

    plt.title(title, fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=8)
    plt.yticks(rotation=0, fontsize=8)
    plt.tight_layout()
    plt.show()
    return CM

def evaluate_model(model_name, y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"\n{model_name}")
    print(f"Accuracy:  {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1 Score:  {f1:.4f}")

    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Diabetes', 'Diabetes'])
    fig, ax = plt.subplots(figsize=(6, 5))
    disp.plot(ax=ax, cmap='Blues')
    ax.set_title(f'{model_name}')
    plt.tight_layout()
    plt.show()

    return {'model': model_name, 'accuracy': accuracy, 'precision': precision, 'recall': recall, 'f1': f1}

def plot_feature_importance(importances, feature_names, model_name, top_n=15):
    feat_imp = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
    feat_imp = feat_imp.sort_values('Importance', ascending=False).head(top_n)

    plt.figure(figsize=(10, 6))
    sns.barplot(data=feat_imp, x='Importance', y='Feature', color='steelblue')
    plt.title(f'Feature Importance - {model_name}')
    plt.tight_layout()
    plt.show()

    return feat_imp

# Data loading and exploration
print("Dataset Information")
print(dataset.head())

print("\nClass Balance (%)")
print(dataset['diagnosed_diabetes'].value_counts(normalize=True) * 100)

print("\nMissing Values")
print(dataset.isnull().sum().sum())
dataset = dataset.dropna()

# Check binary feature distribution
cols = ['cardiovascular_history', 'hypertension_history', 'family_history_diabetes']
for c in cols:
    print(f"\n{c}")
    print(dataset[c].value_counts())

print("\nDataset Statistics")
print(dataset.describe())

# Preprocessing
educational_level_map = {'No formal': 0, 'Highschool': 1, 'Graduate': 2, 'Postgraduate': 3}
income_level_map = {'Low': 0, 'Lower-Middle': 1, 'Middle': 2, 'Upper-Middle': 3, 'High': 4}
dataset['education_level'] = dataset['education_level'].map(educational_level_map)
dataset['income_level'] = dataset['income_level'].map(income_level_map)

# Log transformation for skewed features
log_features = ['alcohol_consumption_per_week', 'physical_activity_minutes_per_week']

for col in log_features:
    plt.figure(figsize=(14, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(dataset[col], kde=True, color='red')
    plt.title(f'Original: {col}')
    plt.xlabel('Value')

    log_transformed = np.log1p(dataset[col])

    plt.subplot(1, 2, 2)
    sns.histplot(log_transformed, kde=True, color='blue')
    plt.title(f'After Log Transformation: {col}')
    plt.xlabel('Log(Value + 1)')

    plt.tight_layout()
    plt.show()

dataset['alcohol_consumption_per_week'] = np.log1p(dataset['alcohol_consumption_per_week'])
dataset['physical_activity_minutes_per_week'] = np.log1p(dataset['physical_activity_minutes_per_week'])

# Correlation matrix BEFORE feature engineering
print("\n" + "="*50)
print("Correlation Matrix - BEFORE Feature Engineering")
print("="*50)
drawCorrelationMatrix(dataset, "Correlation Matrix - Before Feature Engineering")

# Feature engineering
dataset['bmi_age_risk'] = dataset['bmi'] * dataset['age']
dataset['pulse_pressure'] = dataset['systolic_bp'] - dataset['diastolic_bp']
dataset['map_bp'] = dataset['diastolic_bp'] + (dataset['pulse_pressure'] / 3)
dataset['total_hdl_ratio'] = dataset['cholesterol_total'] / (dataset['hdl_cholesterol'] + 1)
dataset['ldl_hdl_ratio'] = dataset['ldl_cholesterol'] / (dataset['hdl_cholesterol'] + 1)
dataset['age_bin'] = pd.qcut(dataset['age'], q=10, labels=False, duplicates='drop')
dataset['bmi_bin'] = pd.qcut(dataset['bmi'], q=10, labels=False, duplicates='drop')

# Correlation matrix AFTER feature engineering (before dropping)
print("\n" + "="*50)
print("Correlation Matrix - AFTER Feature Engineering (before dropping)")
print("="*50)
drawCorrelationMatrix(dataset, "Correlation Matrix - After Feature Engineering")

# Feature and target definition
# Drop features with high multicollinearity and those replaced by engineered features
old_columns_to_drop = [
    'age',                # corr 0.98 with age_bin, 0.90 with bmi_age_risk
    'bmi',                # corr 0.97 with bmi_bin, used in bmi_age_risk
    'bmi_bin',            # low importance (1.9%), redundant
    'age_bin',            # corr 0.88 with bmi_age_risk, lower importance
    'cholesterol_total',  # used to create ratios
    'total_hdl_ratio',    # corr 0.96 with ldl_hdl_ratio, lower importance
    'systolic_bp',        # used to create pulse_pressure
    'diastolic_bp',       # used to create map_bp
]

X = dataset.drop(['diagnosed_diabetes', 'id'] + old_columns_to_drop, axis=1)
y = dataset['diagnosed_diabetes']

# Correlation matrix AFTER dropping features
print("\n" + "="*50)
print("Correlation Matrix - AFTER Dropping Redundant Features")
print("="*50)
X_with_target = X.copy()
X_with_target['diagnosed_diabetes'] = y
drawCorrelationMatrix(X_with_target, "Correlation Matrix - Final (After Dropping)")

target_features = [
    'alcohol_consumption_per_week', 'physical_activity_minutes_per_week',
    'diet_score', 'sleep_hours_per_day', 'screen_time_hours_per_day',
    'waist_to_hip_ratio', 'heart_rate', 'hdl_cholesterol',
    'ldl_cholesterol', 'triglycerides', 'bmi_age_risk',
    'map_bp', 'pulse_pressure', 'ldl_hdl_ratio'
]
# Outlier analysis
print("\n" + "="*50)
print("Outlier Analysis (IQR method)")
print("="*50)

outlier_counts = []
for col in target_features:
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = X[(X[col] < lower) | (X[col] > upper)]
    pct = len(outliers) / len(X) * 100

    if len(outliers) > 0:
        outlier_counts.append({'Feature': col, 'Count': len(outliers), 'Percentage': pct})
        print(f"{col}: {len(outliers)} ({pct:.1f}%)")

print(f"\nNote: Outliers managed through RobustScaler and log transformations")


# Features to one-hot encode
features_to_hot_encoding = ['employment_status', 'smoking_status', 'gender', 'ethnicity']

# ColumnTransformer
Ct = ColumnTransformer(
    [
        ("hot-encoding", OneHotEncoder(drop='first', handle_unknown='ignore'), features_to_hot_encoding),
        ("Scaler", RobustScaler(), target_features)
    ],
    remainder='passthrough'
)

# Train/validation split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=Random_State, stratify=y)
X_train_scaled = Ct.fit_transform(X_train)
X_val_scaled = Ct.transform(X_val)

feature_names = Ct.get_feature_names_out()
print(f"\nTraining set shape: {X_train_scaled.shape}")
print(f"Validation set shape: {X_val_scaled.shape}")

# Apply PCA for dimensionality reduction
print("\nApplying PCA")
pca = PCA(n_components=0.95, random_state=Random_State)
X_train_pca = pca.fit_transform(X_train_scaled)
X_val_pca = pca.transform(X_val_scaled)

print(f"Original dimensions: {X_train_scaled.shape[1]}")
print(f"Reduced dimensions: {X_train_pca.shape[1]}")
print(f"Explained variance: {pca.explained_variance_ratio_.sum():.4f}")

# Storage for results
all_results = []

# Linear models with class_weight
print("\n" + "="*50)
print("Linear Models")
print("="*50)

linear_models = {
    "LogisticRegression": LogisticRegression(max_iter=5000, random_state=Random_State, solver='saga', class_weight='balanced'),
}

for name, model in linear_models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_val_scaled)
    result = evaluate_model(name, y_val, y_pred)
    all_results.append(result)

# Logistic Regression with PCA
print("\n" + "="*50)
print("Logistic Regression with PCA")
print("="*50)

lr_pca = LogisticRegression(max_iter=5000, random_state=Random_State, solver='saga', class_weight='balanced')
lr_pca.fit(X_train_pca, y_train)
y_pred = lr_pca.predict(X_val_pca)
result = evaluate_model("LogisticRegression (PCA)", y_val, y_pred)
all_results.append(result)

# Random Forest with class_weight
print("\n" + "="*50)
print("Random Forest")
print("="*50)

n_estimators_list = [50, 100 ,150, 200]

for n in n_estimators_list:
    rf = RandomForestClassifier(n_estimators=n, n_jobs=-1, max_depth=12, min_samples_leaf=5,
                                 random_state=Random_State, class_weight='balanced')
    rf.fit(X_train_scaled, y_train)
    y_pred_rf = rf.predict(X_val_scaled)

    result = evaluate_model(f"RandomForest (n={n})", y_val, y_pred_rf)
    all_results.append(result)

    feat_imp_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': rf.feature_importances_
    }).sort_values('Importance', ascending=False)

    print(f"\nFeature Importance (n_estimators={n})")
    print(feat_imp_df.head(15).to_string(index=False))

    if n == n_estimators_list[-1]:
        plot_feature_importance(rf.feature_importances_, feature_names, f"RandomForest (n={n})")
print("\n" + "="*50)
print("Random Forest - Tree Visualization")
print("="*50)

from sklearn.tree import plot_tree

# Prendi il primo albero (o uno a caso)
sample_tree = rf.estimators_[0]

plt.figure(figsize=(20, 10))
plot_tree(sample_tree,
          feature_names=feature_names,
          class_names=['No Diabetes', 'Diabetes'],
          filled=True,
          max_depth=3,  # Limita profondit per leggibilit
          fontsize=10,
          rounded=True)
plt.title("Sample Decision Tree from Random Forest (depth limited to 3)")
plt.tight_layout()
plt.show()

print(f"Tree depth: {sample_tree.get_depth()}")
print(f"Number of leaves: {sample_tree.get_n_leaves()}")
# XGBoost with scale_pos_weight
print("\n" + "="*50)
print("XGBoost")
print("="*50)

# Calculate scale_pos_weight for imbalanced data
neg_count = (y_train == 0).sum()
pos_count = (y_train == 1).sum()
scale_pos_weight = neg_count / pos_count

xgb = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos_weight,
    random_state=Random_State,
    n_jobs=-1,
    eval_metric='logloss'
)

xgb.fit(X_train_scaled, y_train)
y_pred_xgb = xgb.predict(X_val_scaled)

result = evaluate_model("XGBoost", y_val, y_pred_xgb)
all_results.append(result)

xgb_feat_imp_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': xgb.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nFeature Importance (XGBoost)")
print(xgb_feat_imp_df.head(15).to_string(index=False))
plot_feature_importance(xgb.feature_importances_, feature_names, "XGBoost")

# Create BALANCED subset for KNN and SVM
print("\n" + "="*50)
print("Creating Balanced Subset for KNN and SVM")
print("="*50)

n_per_class = 10000

# Create temporary DataFrame for easy sampling
train_df = pd.DataFrame(X_train_scaled)
train_df['target'] = y_train.values

# Sample balanced
df_healthy = train_df[train_df['target'] == 0].sample(n=n_per_class, random_state=Random_State)
df_diabetic = train_df[train_df['target'] == 1].sample(n=n_per_class, random_state=Random_State)
df_balanced = pd.concat([df_healthy, df_diabetic]).sample(frac=1, random_state=Random_State)

X_balanced = df_balanced.drop('target', axis=1).values
y_balanced = df_balanced['target'].values

print(f"Balanced subset: {len(X_balanced)} samples")
print(f"Class distribution: {np.bincount(y_balanced.astype(int))}")

# Same for PCA
train_df_pca = pd.DataFrame(X_train_pca)
train_df_pca['target'] = y_train.values

df_healthy_pca = train_df_pca[train_df_pca['target'] == 0].sample(n=n_per_class, random_state=Random_State)
df_diabetic_pca = train_df_pca[train_df_pca['target'] == 1].sample(n=n_per_class, random_state=Random_State)
df_balanced_pca = pd.concat([df_healthy_pca, df_diabetic_pca]).sample(frac=1, random_state=Random_State)

X_balanced_pca = df_balanced_pca.drop('target', axis=1).values
y_balanced_pca = df_balanced_pca['target'].values

# KNN without PCA (balanced)
print("\n" + "="*50)
print("KNN without PCA (balanced 20k subset)")
print("="*50)

k_values = [5, 50, 100, 141, 200]

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    knn.fit(X_balanced, y_balanced)
    y_pred = knn.predict(X_val_scaled)
    result = evaluate_model(f"KNN (k={k}, balanced)", y_val, y_pred)
    all_results.append(result)

# KNN with PCA (balanced)
print("\n" + "="*50)
print("KNN with PCA (balanced 20k subset)")
print("="*50)

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    knn.fit(X_balanced_pca, y_balanced_pca)
    y_pred = knn.predict(X_val_pca)
    result = evaluate_model(f"KNN (k={k}, PCA, balanced)", y_val, y_pred)
    all_results.append(result)

# SVM without PCA (balanced)
print("\n" + "="*50)
print("SVM without PCA (balanced 20k subset)")
print("="*50)

svm_kernels = ['rbf', 'poly']

for k in svm_kernels:
    svm_model = SVC(kernel=k, C=1.0, gamma='scale', class_weight='balanced')
    svm_model.fit(X_balanced, y_balanced)
    y_pred = svm_model.predict(X_val_scaled)
    result = evaluate_model(f"SVM ({k}, balanced)", y_val, y_pred)
    all_results.append(result)

# SVM with PCA (balanced)
print("\n" + "="*50)
print("SVM with PCA (balanced 20k subset)")
print("="*50)

for k in svm_kernels:
    svm_model = SVC(kernel=k, C=1.0, gamma='scale', class_weight='balanced')
    svm_model.fit(X_balanced_pca, y_balanced_pca)
    y_pred = svm_model.predict(X_val_pca)
    result = evaluate_model(f"SVM ({k}, PCA, balanced)", y_val, y_pred)
    all_results.append(result)

# Cross-validation on best models
print("\n" + "="*50)
print("Cross-Validation (StratifiedKFold, 5 folds)")
print("="*50)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=Random_State)

cv_models = {
    "LogisticRegression": LogisticRegression(max_iter=5000, random_state=Random_State, solver='saga', class_weight='balanced'),
    "RandomForest": RandomForestClassifier(n_estimators=100, n_jobs=-1, max_depth=12, min_samples_leaf=5,
                                           random_state=Random_State, class_weight='balanced'),
    "XGBoost": XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.1, subsample=0.8,
                             colsample_bytree=0.8, scale_pos_weight=scale_pos_weight,
                             random_state=Random_State, n_jobs=-1, eval_metric='logloss'),
}

cv_results = {}
for name, model in cv_models.items():
    scores = cross_val_score(model, X_train_scaled, y_train, cv=cv, scoring='f1')
    cv_results[name] = scores
    print(f"{name}")
    print(f"F1 scores: {scores.round(4)}")
    print(f"Mean F1: {scores.mean():.4f} (+/- {scores.std():.4f})")

plt.figure(figsize=(10, 5))
plt.boxplot([cv_results[name] for name in cv_results], labels=cv_results.keys())
plt.ylabel('F1 Score')
plt.title('Cross-Validation F1 Score Comparison')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# ROC Curve and AUC
print("\n" + "="*50)
print("ROC Curve and AUC")
print("="*50)

roc_models = {
    "LogisticRegression": LogisticRegression(max_iter=5000, random_state=Random_State, solver='saga', class_weight='balanced'),
    "RandomForest": RandomForestClassifier(n_estimators=100, n_jobs=-1, max_depth=12, min_samples_leaf=5,
                                           random_state=Random_State, class_weight='balanced'),
    "XGBoost": XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.1, subsample=0.8,
                             colsample_bytree=0.8, scale_pos_weight=scale_pos_weight,
                             random_state=Random_State, n_jobs=-1, eval_metric='logloss'),
}

plt.figure(figsize=(10, 8))

for name, model in roc_models.items():
    model.fit(X_train_scaled, y_train)
    y_pred_proba = model.predict_proba(X_val_scaled)[:, 1]

    fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, lw=2, label=f'{name} (AUC = {roc_auc:.4f})')
    print(f"{name} - AUC: {roc_auc:.4f}")

plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Classifier (AUC = 0.5)')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Final comparison table
print("\n" + "="*50)
print("Final Model Comparison")
print("="*50)

results_df = pd.DataFrame(all_results)
results_df = results_df.sort_values('f1', ascending=False).reset_index(drop=True)
print(results_df.to_string(index=False))

# Plot final comparison
fig, ax = plt.subplots(figsize=(14, 6))
metrics = ['accuracy', 'precision', 'recall', 'f1']
x = np.arange(len(results_df))
width = 0.2

for i, metric in enumerate(metrics):
    ax.bar(x + i * width, results_df[metric], width, label=metric.capitalize())

ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(results_df['model'], rotation=45, ha='right')
ax.set_ylabel('Score')
ax.set_title('Model Comparison')
ax.legend()
ax.set_ylim(0.0, 1.0)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
